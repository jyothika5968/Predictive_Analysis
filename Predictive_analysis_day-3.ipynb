```python
#predictive analysis
#Smart water tank
water_consumption=[20,10,40,30,50]
days=[2,4,3,5,6]

#find:
#avg of water
#total water usage in a day
#water usage increased
# prediction
#sum of daily increased water
#shortage will occur or not
```

```python
#avaerage of water
avg_water=sum(water_consumption)/len(water_consumption)
print(avg_water)
```

```python
#total water usage in a day
total_water=sum(water_consumption)
print(total_water)
```

```python
#water usage increased
water_increased=water_consumption[3]-water_consumption[1]
print(water_increased)
```

```python
# method 1
daily_increased_water = sum(water_consumption[1:])- sum(water_consumption[:-1])
print(daily_increased_water)
```

```python
#Prediction
prediction=avg_water+water_increased
print(prediction)
```

```python
# shortage will occur or not
if daily_increased_water < avg_water:
  print("Shortage will occur")
else:
  print("Shortage will not occur")
```
